# EDA

```python
from IPython.display import Audio
import matplotlib.pyplot as plt
import librosa
import numpy as np

samples = [
    "./data/visual/3d86b69a_nohash_0.wav",
    "./data/dog/881583a6_nohash_0.wav",
    "./data/zero/4fd1443e_nohash_2.wav"
]

ram_sample = ["./ram_data/ram_bed_1.wav",
             "./ram_data/ram_bed_2.wav",
              "./ram_data/ram_backward_1.wav",
              "./ram_data/ram_visual_1.wav",
              "./ram_data/shyam_forward_1.wav"
             ]

def show_melspec(audio_path):
    print(f"Showing {audio_path}")
    audio_settings = {
        "sr": 16000,
        "n_mels": 40,
        "n_fft": 480,
        "win_length": 480,
        "hop_length": 160,
        "center": False
    }
    x, sr = librosa.load(audio_path, sr=audio_settings["sr"])
    x = librosa.util.fix_length(x, size=audio_settings["sr"])
    x = librosa.feature.melspectrogram(y=x, **audio_settings)  # Remove `sr=sr` here
  
    plt.figure(figsize=(7, 5))
    plt.imshow(librosa.power_to_db(x, ref=np.max), cmap="hot", aspect="auto")
    plt.title(f'Mel-spectrogram of {audio_path}')
    plt.xlabel('Time')
    plt.ylabel('Mel-frequency')
    plt.colorbar(format='%+2.0f dB')
    plt.show()
```

```python

# Display mel-spectrogram and play audio
show_melspec(samples[0])
Audio(samples[0])
```

```python
show_melspec(samples[1])
Audio(samples[1])
```

```python
show_melspec(samples[2])
Audio(samples[2])
```
